Q1. 주어진 자연수가 홀수인지 짝수인지 판별해주는 함수를 작성하시오.

결과 👇

(홀수 입력시) 주어진 자연수는 홀수입니다.

(짝수 입력시) 주어진 자연수는 짝수입니다.


```python
#Q1. 
# 1. 함수 정의
def is_Odd(num):
  if num % 2 == 0 :
    print("주어진 자연수는 짝수입니다.")
  else:
    print("주어진 자연수는 홀수입니다.")


# 2. 사용자에게 입력 받음
user = int(input("숫자를 입력해주세요 >> ")) #사용자에게 입력받은 숫자 저장

# 3. 함수에 입력값 넣기
is_Odd(user) # 호출
```

Q3. 가위, 바위, 보를 입력받아 승패를 결정하는 코드를 작성하시오.

R_s_p = ['가위', '바위', '보']

결과 👇

(가위로 이겼을 때) 가위가 이겼습니다.

(바위로 이겼을 때) 바위가 이겼습니다.

(보로 이겼을 때) 보가 이겼습니다.

(비겼을 때) 비겼습니다.

```python
# Q3.

# 리스트 정의
R_s_p = ['가위', '바위','보']

#1. 함수 정의
def check(a,b):
  if a==b:
    print("비겼습니다!")
  elif(a==R_s_p[0] and b==R_s_p[1]) or (a==R_s_p[1] and b==R_s_p[0]):
    print("바위가 이겼습니다!")
  elif(a==R_s_p[0] and b==R_s_p[2]) or (a==R_s_p[2] and b==R_s_p[0]):
    print("가위가 이겼습니다!")
  elif(a==R_s_p[1] and b==R_s_p[2]) or (a==R_s_p[2] and b==R_s_p[1]):
    print("보가 이겼습니다!")

# 사용자에게 값 2개 받기
user1 = input("user 1 >> ")
user2 = input("user 2 >> ")

# 함수에 user1, user2 값 넣기
check(user1, user2)
```

```python
# Q2. 입력값이 정해져 있지 않을 때 평균값 계산 

# 총합, 입력값의 개수

def avg_numbers(*numbers):
  sum = 0
  for i in numbers :  
    sum += i
  avg = sum / len(numbers)
  return avg


print(avg_numbers(1,2,3)) # 2
print(avg_numbers(1,2,3,4,5)) # 3
```

```python
# 재귀함수
# 함수가 간접적 or 직접적으로 자신을 호출하는 프로세스


# 팩토리얼 : 10! = 10x9x8x7x6x5x4x3x2x1
#             5! = 5x4x3x2x1

user = int(input("숫자를 입력해주세요 >> ")) #10

def factorial(num) : # num = 10 9 8 .... 0
  if num == 0 : # 0! = 1
    return 1 # 재귀함수 x
  else :
    return num * factorial(num-1) # 10 * 9

print(factorial(user))

```

```python
# 피보나치 수열 (1번째 : 1) (2번째 : 1) 2 3 5 8 13
def fibo(num) : # 몇 번째 자리인지
  if num <= 1:
    return num
  return fibo(num-1) +fibo(num-2)

print(fibo(6))
```
