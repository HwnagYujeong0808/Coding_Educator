```python
product = {'ÏÉùÏàò':800, 'Ï¥àÏΩúÎ¶ø':1000, 'Îß•Ï£º':2500, 'Í∑§':5000}
def Market(a):
  for key, value in product.items():
    if a == key: # ÏûÖÎ†•Ìïú Í∞í == ÎîïÏÖîÎÑàÎ¶¨ ÏïàÏùò keyÍ∞í
      print("{}Ïùò Í∞ÄÍ≤©ÏùÄ {}ÏõêÏûÖÎãàÎã§.".format(key, value))
    else:  
      break
user = input("ÏÇ¨Í≥† Ïã∂ÏùÄ Î¨ºÍ±¥ÏùÑ ÏûÖÎ†•Ìï¥Ï£ºÏÑ∏Ïöî >> ")
Market(user)

```

```python
# Q2. ÏûÖÎ†•Í∞íÏù¥ Ï†ïÌï¥Ï†∏ ÏûàÏßÄ ÏïäÏùÑ Îïå ÌèâÍ∑†Í∞í Í≥ÑÏÇ∞ 

# Ï¥ùÌï©, ÏûÖÎ†•Í∞íÏùò Í∞úÏàò

def avg_numbers(*numbers):
  sum = 0
  for i in numbers :  
    sum += i
  avg = sum / len(numbers)
  return avg


print(avg_numbers(1,2,3)) # 2
print(avg_numbers(1,2,3,4,5)) # 3
```

```python
# Ïû¨Í∑ÄÌï®Ïàò
# Ìï®ÏàòÍ∞Ä Í∞ÑÏ†ëÏ†Å or ÏßÅÏ†ëÏ†ÅÏúºÎ°ú ÏûêÏã†ÏùÑ Ìò∏Ï∂úÌïòÎäî ÌîÑÎ°úÏÑ∏Ïä§


# Ìå©ÌÜ†Î¶¨Ïñº : 10! = 10x9x8x7x6x5x4x3x2x1
#             5! = 5x4x3x2x1

user = int(input("Ïà´ÏûêÎ•º ÏûÖÎ†•Ìï¥Ï£ºÏÑ∏Ïöî >> ")) #10

def factorial(num) : # num = 10 9 8 .... 0
  if num == 0 : # 0! = 1
    return 1 # Ïû¨Í∑ÄÌï®Ïàò x
  else :
    return num * factorial(num-1) # 10 * 9

print(factorial(user))

```

```python
# ÌîºÎ≥¥ÎÇòÏπò ÏàòÏó¥ (1Î≤àÏß∏ : 1) (2Î≤àÏß∏ : 1) 2 3 5 8 13
def fibo(num) : # Î™á Î≤àÏß∏ ÏûêÎ¶¨Ïù∏ÏßÄ
  if num <= 1:
    return num
  return fibo(num-1) +fibo(num-2)

print(fibo(6))
```
