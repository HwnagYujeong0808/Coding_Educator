```python
# while 문의 continue (중간에서 멈추고 처음으로 돌아가라)

#10보다 작은 홀수만 출력
a = 0
result = 0
while a < 10 :
  a += 1
  if a % 2 ==0 : #짝수가 될 경우
    continue # a가 짝수일 때 아래 문장 수행 x 
  print(a)
  result += a

print("결과 > ", result)  
```

```python
# 책 예제
# 커피, 거스름돈 계산

coffee = 10
while True :
  money = int(input("돈을 넣어주세요 : "))
  if money == 300 :                                 # if 거스름돈이 없는 경우
    print("커피를 받아가세요~")
    coffee -= 1  # coffee = coffee - 1
  elif money > 300 :                                # elif 거스름돈이 있는 경우
    print("거스름돈 %d원을 받아가세요~"%(money - 300))
    print("커피를 받아가세요~")
    coffee -= 1
  else :                                             # else 돈이 부족한 경우
    print("커피를 살 수 없습니다.")  
  if coffee == 0:
    print("커피가 다 떨어져 살 수 없습니다. 판매를 중지합니다.")
    break

    
```

```python
# 구구단 
# 2단부터 9단까지 출력
# 입력받은 단 ~ 9단
user = int(input("몇 단? ")) 

while user < 10 :
  print("============%d단==========="%user)
  j=1
  while j < 10 :
    print("{} * {} = {}".format(user, j, user * j))
    j = j + 1
  user += 1   


#중첩 반복문
# while문 안에 while문
# 
# while [조건]:
#   [수행하고 싶은 문장 1]
#   while [조건2] :
#     [수행하고 싶은 문장 2]
```
