```python
#집합

#1. 교집합 : &

s1 = set([1,2,3,4,5])
s2 = set([4,5,6,7,8])

# 결과 : {4,5}

print(s1 & s2)

# 교집합 메서드 : .intersection

print(s1.intersection(s2))

#2. 합집합 : |
# 결과 : {1,2,3,4,5,6,7,8}
print(s1 | s2)

# 합집합 메서드(함수) .함수명() :   .union() 
print(s1.union(s2))

# 3. 차집합 : -
# s1 - s2 결과 : {1,2,3}
print(s1- s2)

# 차집합 메서드(함수) : difference  
# 차집합에서는 s1, s2의 순서가 매우 중요함
print(s1.difference(s2)) # {1,2,3}
print(s2.difference(s1)) # {8,6,7}


#집합 관련 메소드 add(), update()
s1 = set([1,2,3,4,5])
s2 = set([4,5,6,7,8])
s3 = {1,2,3,4,5,6,7,8}

# add는 값 하나 추가할 때
s1.add(6)
print(s1) #{1,2,3,4,5,6}

#update는 값 여러 개 추가할 때
s1.update([9,10,11])
print(s1)
print(s3)

# remove 함수 = discard
s1.remove(3)
print(s1) # 3 빼고 출력됨

s1.discard(5)
print(s1)
```
