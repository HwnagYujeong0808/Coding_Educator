```python
# 메서드 오버라이딩

# 똑같은 함수(메서드)의 속성을 변경

# 똑같은 함수명을 사용하고, 자식클래스에서 동일 함수를 다시 만드는 것

class calculator :
  def __init__(self, num1,  num2): #초깃값 생성 (생성자)
    self.num1 = num1 
    self.num2 = num2
  def sum(self):
    result = self.num1 + self.num2
    return result
  def sub(self):
    result = self.num1 - self.num2
    return result
  def mult(self):
    result = self.num1 * self.num2
    return result
  def div(self):
    result = self.num1 / self.num2
    return result

  class OverrideCalculator(calculator) :
    def div(self) : #메서드 오버라이딩에서 함수명 동일
      if self.num2 == 0:
        print("나눗셈이 불가능합니다.")
      else :
        print("결과는 {}입니다.".format(self.num1 / self.num2))

  calcul1 = OverrideCalculator(10, 0)
  calcul1.div() # 나눗셈이 불가능합니다. = > 메서드 오버라이딩이 발생
  # 부모 클래스의 함수(메서드) 대신 오버라이딩한 함수(메서드)가 호출됨
          
```

```python
# 원의 넓이, 원의 둘레 길이 구하는 클래스

import math # math.pi (원주율)

class Circle : 
  def __init__(self, radius):
    self.radius =radius  # 반지름
  def circleRound(self) : # 원의 둘레
    return 2 * self.radius * math.pi 
  def circleArea(self) : # 원의 넓이
    return self.radius ** 2 * math.pi
circle1 = Circle(5)
print(circle1.circleRound()) # 31.41592653589793
print(circle1.circleArea())  # 78.53981633974483
```

```python
# 클래스 복습 #

class Calculate : # 클래스 정의 
  def __init__(self, num1, num2) : # 초기 생성자 정의
    self.num1 = num1
    self.num2 = num2
  def add(self) :
    return self.num1 + self.num2 
  def minus(self):
    return self.num1 - self.num2

C1 = Calculate(4,2)
C1.add() # 4 + 2
C1.minus() # 4 - 2

# 클래스 상속 : 부모클래스(더하기 + 빼기 ) + (곱하기, 나누기)

class Calculate2(Calculate) : # 초기 생성자 정의된 상태(부모클래스에서)
  def multiple(self) :
    return self.num1 * self.num2
  def devide(self):
    return self.num1 / self.num2

C2 = Calculate2(4,2)
print(C2.add()) # 6
print(C2.minus()) # 2
print(C2.multiple()) # 8
print(C2.devide()) # 2.0

# 메서드 오버라이딩 : 부모클래스에서 정의된 함수의 내용을 바꿔주는 것

# greeting 클래스

class greeting :
  def __init__(self):
    print("안녕하세요")
person1 = greeting()
print(person1) # 안녕하세요

class greeting_school(greeting):
  def __init__(self): # 메서드 오버라이딩
    print("안녕하세요! 저는 학생입니다.")

person2 = greeting_school()
print(person2)  # 안녕하세요! 저는 학생입니다  

```
