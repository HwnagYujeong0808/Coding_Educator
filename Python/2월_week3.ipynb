```python
 # 예외처리 

 # 4 / 0
 # a = [1,2,3,4,5] a[6]

'''
try :
  실행할 코드
except :
  예외가 발생했을 경우 처리할 코드  

'''

x = int(input("숫자를 입력해주세요 >> ")) # 0 입력 시 ZeroDivisionError 발생
try :
  y = 10 / x
  print(y)
except :
  print("에러가 발생했습니다.")
```

```python
# 인덱싱 에러 : ex) [0,1,2,3] 5번째 인덱스 호출할 때 발생하는 에러

a = [1,2,3,4,5]
try :
  print(4/0) # ZeroDivisionError 발생
  print(a[10]) # IndexError 발생
except IndexError:
  print("인덱싱 불가!")  # 인덱싱 불가!
except ZeroDivisionError :
  print("0으로 나눌 수 없음!")  
  
```

```python
''' 예외처리의 기본 구조
try :
  실행 코드
except :
  예외가 발생하였을 때 수행할 코드
else :
  예외가 발생x 수행할 코드
finally :    
  예외발생여부와 관계 없이 항상 수행할 코드
'''

# 닉네임을 입력하는 프로그램 
# 만약, 바보 입력시 '허용되지 않는 별명입니다' 문구 출력

# 1 . 에러 발생 클래스 생성
class MyError(Exception) :
  pass

# 2. 별명을 출력해주는 함수 작성
def nickName(name) :
  if name == "바보" :
    raise MyError() # 에러 발생
  print(name)  
try :  
  nickName("천사")
  nickName("바보")
  
except MyError:
  print("허용되지 않는 별명입니다") 
```

```python
# 내장 함수

# abs() : 절댓값
abs(-10) # 10

# all() : 자료형 x가 모두 참이면 True, 거짓이 하나라도 있으면 False
all([1,2,3]) # 0이 포함되어 있으면 거짓 리턴
# True

all([0,1,1,1]) # False

# any() : 자료형 x가 하나라도 참이면 True, 모두 거짓일 때만 False
any([0,1,2,3]) # True
any([0,0,0,0]) # False

# chr() : 숫자 입력받아 해당 문자 출력 (아스키코드 : 문자를 숫자로 바꾼 값)
# a : 97
# A : 65
chr(97) # a
chr(65) # A

# divmod(a,b) : a를 b로 나눈 몫과 나머지 리턴
divmod(7,4) # (1,3)

# enumerate : for 문에서 사용 
# 인덱스와 요소 이름을 포함하는 객체를 돌려줌 : 변수로 (i, name) 함께 사용
for i, name in enumerate(['strawberry', 'grape', 'banana']):
  print(i, name)
# 인덱스 시작을 0이 아닌 1로 바꿀 수 있음
for i, name in enumerate(['strawberry', 'grape', 'banana'], start = 1):
  print(i, name) 
```
